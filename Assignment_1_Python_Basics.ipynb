```python
# Task: Deduct withdrawal amount if balance is sufficient

balance = float(input("Enter Balance: "))
withdrawal = float(input("Enter Withdrawal: "))

if balance >= withdrawal:
    balance -= withdrawal
    print(f"Transaction successful. Remaining balance: ₹{balance}")
else:
    print("Error: Insufficient balance.")
```

```python
# Task: Calculate bill based on tiered pricing [cite: 16]
units = int(input("Units consumed: "))
bill = 0

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + (units - 100) * 7
else:
    bill = (100 * 5) + (100 * 7) + (units - 200) * 10

print(f"Bill amount = {bill}")
```

```python
# Task: Convert between Celsius and Fahrenheit [cite: 26]
temp = float(input("Temp = "))
unit = input("Unit (C/F) = ").upper()

if unit == "C":
    fahrenheit = (temp * 9/5) + 32
    print(f"Temperature in Fahrenheit: {fahrenheit}°F") [cite: 36]
elif unit == "F":
    celsius = (temp - 32) * 5/9
    print(f"Temperature in Celsius: {celsius}°C")
```
